```python
# -*- coding: utf-8 -*-
"""
数据预处理代码
基于原始数据文件：D:\Desktop\2026春季《数据挖掘》考试规定动作 + 基础数据.xlsx
输出：招聘数据_预处理后.xlsx
"""

import pandas as pd
import numpy as np
import re

# ============================================================
# 1. 数据加载
# ============================================================
file_path = r'D:\Desktop\2026春季《数据挖掘》考试规定动作 + 基础数据.xlsx'
df = pd.read_excel(file_path, header=0)  # 以第一行作为列名

print(f"原始数据行数：{len(df)}")
print(f"原始数据列名：{df.columns.tolist()}")
print("\n数据前3行预览：")
print(df.head(3))

# 根据实际列名重命名（原始数据列名为：标题、标题1、标题2、标题3、标题4、标题5、内容）
# 由于您的数据中列名是"标题"系列，但部分数据已有表头，需根据实际情况调整
# 若第一行已经是列名（岗位全称、部门等），则无需重命名
# 检查第一行是否包含"岗位全称"等关键词
if '岗位全称' not in df.columns:
    # 若没有，则手动设置列名
    df.columns = ['岗位全称', '部门', '工作地点', '岗位类别', '招聘人数', '发布日期', '内容']
else:
    print("列名已正确识别")

print(f"\n当前列名：{df.columns.tolist()}")

# ============================================================
# 2. 数据清洗
# ============================================================

# 2.1 重复值处理
# 依据“岗位编号+发布日期”去重，保留最新发布的一条记录
# 从“岗位全称”中提取岗位编号，如“数据工艺（J73201）” -> “J73201”
def extract_job_id(job_title):
    if pd.isna(job_title):
        return ''
    match = re.search(r'（(J\d+)）', str(job_title))
    return match.group(1) if match else ''

df['岗位编号'] = df['岗位全称'].apply(extract_job_id)
print(f"\n提取岗位编号完成，共提取 {df['岗位编号'].nunique()} 个唯一编号")

# 发布日期转换
df['发布日期'] = pd.to_datetime(df['发布日期'])
df = df.sort_values('发布日期', ascending=False)
df = df.drop_duplicates(subset=['岗位编号'], keep='first')
print(f"去重后数据行数：{len(df)}")

# 2.2 缺失值处理
key_fields = ['岗位全称', '岗位类别', '内容']
for field in key_fields:
    missing_count = df[field].isnull().sum()
    print(f"字段 '{field}' 缺失数量：{missing_count}")

# 岗位描述缺失的处理
df['内容'] = df['内容'].fillna('')
# 剔除岗位名称和岗位描述都缺失的记录
df = df[~((df['内容'] == '') & (df['岗位全称'].isnull()))]

# 部门字段标注为"Unknown"的记录，归类至"其他"
df['部门'] = df['部门'].replace('Unknown', '其他')

# 2.3 异常值处理
# 检查发布日期年份是否在合理范围（2021-2026）
df['年份'] = df['发布日期'].dt.year
df = df[(df['年份'] >= 2021) & (df['年份'] <= 2026)]

# 招聘人数处理：提取数值
def extract_people_count(val):
    if pd.isna(val):
        return 1
    if isinstance(val, (int, float)):
        return int(val)
    # 提取数字，如 "2人" -> 2, "若干" -> 1
    match = re.search(r'(\d+)', str(val))
    return int(match.group(1)) if match else 1

df['招聘人数_数值'] = df['招聘人数'].apply(extract_people_count)
df = df[df['招聘人数_数值'] > 0]

# 剔除岗位类别为空或无效的记录
df = df[df['岗位类别'].notna()]

print(f"清洗后数据行数：{len(df)}")

# ============================================================
# 3. 字段标准化
# ============================================================

# 3.1 日期字段处理
df['发布年份'] = df['发布日期'].dt.year
df['发布季度'] = df['发布日期'].dt.quarter
df['发布月份'] = df['发布日期'].dt.month
print("日期字段处理完成")

# 3.2 岗位类别编码
category_mapping = {
    '技术': 3,
    '产品': 2,
    '政企行业解决方案和服务': 4,
    '专业服务和管理支持': 1
}
df['岗位类别_encoded'] = df['岗位类别'].map(category_mapping).fillna(0).astype(int)
print("岗位类别编码完成")

# 3.3 部门字段标准化
df['一级部门'] = df['部门'].str.replace(r'\(.*\)', '', regex=True).str.strip()
print("部门字段标准化完成")

# 3.4 工作地点处理
def extract_main_city(location):
    if pd.isna(location):
        return '未知'
    cities = str(location).split(',')
    return cities[0].strip()

df['主要城市'] = df['工作地点'].apply(extract_main_city)
print("工作地点处理完成")

# ============================================================
# 4. 岗位描述文本清洗
# ============================================================

# 4.1 文本清洗
def clean_text(text):
    if pd.isna(text):
        return ''
    # 去除HTML标签
    text = re.sub(r'<[^>]+>', '', str(text))
    # 去除换行符等特殊字符
    text = re.sub(r'[\r\n\t]+', ' ', text)
    # 去除多余空白
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['清洗后描述'] = df['内容'].apply(clean_text)
print("文本清洗完成")

# 4.2 分词处理（需要安装jieba，若未安装请取消注释安装命令）
# !pip install jieba -q
try:
    import jieba
    import jieba.analyse
    
    # 加载停用词表（若文件不存在则使用默认停用词集）
    def load_stopwords(filepath='stopwords.txt'):
        try:
            with open(filepath, 'r', encoding='utf-8') as f:
                stopwords = set([line.strip() for line in f])
            return stopwords
        except FileNotFoundError:
            return set(['的', '了', '在', '是', '我', '有', '和', '就', '不', '人', '都', '一', '一个', '上', '也', '很', '到', '说', '要', '去', '你', '会', '着', '没有', '看', '好', '自己', '这', '那', '它', '他', '她', '们'])
    
    stopwords = load_stopwords()
    
    def segment_text(text):
        words = jieba.cut(text)
        result = [w for w in words if w not in stopwords and len(w) > 1 and not w.isdigit()]
        return ' '.join(result)
    
    df['分词后描述'] = df['清洗后描述'].apply(segment_text)
    print("分词处理完成")
    
except ImportError:
    print("注意：jieba未安装，跳过分词处理，使用清洗后文本")
    df['分词后描述'] = df['清洗后描述']

# ============================================================
# 5. 技能标签提取
# ============================================================

SKILL_KEYWORDS = {
    'Skill_Python': r'python',
    'Skill_SQL': r'SQL|sql|数据库查询',
    'Skill_Java': r'java',
    'Skill_Cpp': r'C\+\+|cpp',
    'Skill_Go': r'Go|Golang|golang',
    'Skill_LLM': r'大模型|LLM|文心|大语言模型',
    'Skill_ML': r'机器学习|深度学习|machine learning|deep learning',
    'Skill_DM': r'数据挖掘|挖掘算法',
    'Skill_DA': r'数据分析|统计分析',
    'Skill_BigData': r'大数据|Hadoop|Spark|Flink',
    'Skill_DB': r'数据库|MySQL|Redis|MongoDB',
    'Skill_DataWarehouse': r'数据仓库|ETL|数仓',
    'Skill_Cloud': r'云原生|容器|Kubernetes|Docker|K8s',
    'Skill_Frontend': r'前端|Web|React|Vue|Angular',
    'Skill_PM': r'项目管理|PMO'
}

def extract_skills(text):
    """从文本中提取技能标签"""
    skills = {}
    for skill_name, pattern in SKILL_KEYWORDS.items():
        if re.search(pattern, str(text), re.IGNORECASE):
            skills[skill_name] = 1
        else:
            skills[skill_name] = 0
    return skills

# 批量提取技能标签
skill_df = df['清洗后描述'].apply(extract_skills).apply(pd.Series)
df = pd.concat([df, skill_df], axis=1)

# 计算技能总数
skill_cols = list(SKILL_KEYWORDS.keys())
df['技能总数'] = df[skill_cols].sum(axis=1)
print("技能标签提取完成")

# ============================================================
# 6. 输出统计
# ============================================================
print(f"\n最终有效样本数：{len(df)}")
print(f"最终字段数：{len(df.columns)}")
print("\n各技能提及次数：")
for skill in skill_cols:
    count = df[skill].sum()
    print(f"  {skill}: {count} ({count/len(df)*100:.1f}%)")

# ============================================================
# 7. 保存预处理后的数据
# ============================================================
output_path = r'D:\Desktop\招聘数据_预处理后.xlsx'
df.to_excel(output_path, index=False)
print(f"\n预处理完成！数据已保存到：{output_path}")
```

```python
# 7. 技能需求的时间演变趋势（表5）
# 按年份统计各项技能提及率
yearly_skill = df.groupby('年份')[skill_cols].mean() * 100  # 转为百分比
yearly_skill = yearly_skill.round(1)

# 选取核心技能：LLM、Python、ML、DA、云原生、SQL、Java
core_skills = ['Skill_LLM', 'Skill_Python', 'Skill_ML', 'Skill_DA', 'Skill_Cloud', 'Skill_SQL', 'Skill_Java']
table5 = yearly_skill[core_skills].T

# 计算增长幅度（使用数据中实际存在的年份）
years = sorted(table5.columns)  # 获取所有年份（这里是列名）
if len(years) >= 2:
    start_year = years[0]
    end_year = years[-1]
    growth = (table5[end_year] - table5[start_year]).round(1)
    # 将增长幅度作为新列添加到 table5 中（注意列名）
    table5['增长幅度'] = growth
else:
    print("警告：数据中年份不足两年，无法计算增长幅度")

print("\n表5 核心技能提及率的年度变化")
print(table5)

# 绘制技能演变折线图（只使用实际年份列，不含“增长幅度”）
plt.figure(figsize=(12, 7))
for skill in core_skills:
    # 只取年份列（排除“增长幅度”）
    plt.plot(years, table5.loc[skill, years], marker='o', label=skill.replace('Skill_', ''))
plt.xlabel('年份', fontsize=12)
plt.ylabel('提及率（%）', fontsize=12)
plt.title('核心技能提及率年度变化趋势', fontsize=14)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('技能演变趋势.png', dpi=300)
plt.show()
```

```python
# ============================================================
# 8. 不同岗位类别的技能需求差异（表6）
# ============================================================
# 岗位类别编码映射
cat_mapping = {3: '技术类', 2: '产品类', 4: '政企解决方案类', 1: '专业服务类'}
df['岗位类别名称'] = df['岗位类别_encoded'].map(cat_mapping)

category_skill = df.groupby('岗位类别名称')[skill_cols].mean() * 100
category_skill = category_skill.round(1)

# 添加平均技能数
avg_skills = df.groupby('岗位类别名称')['技能总数'].mean().round(1)
category_skill['技能总数（平均）'] = avg_skills

# 选取部分技能展示（与表6对应）
display_skills = ['Skill_Python', 'Skill_SQL', 'Skill_Java', 'Skill_LLM', 'Skill_ML', 
                  'Skill_DA', 'Skill_Cloud', 'Skill_BigData']
table6 = category_skill[display_skills + ['技能总数（平均）']]
# 重命名列
table6.columns = ['Python', 'SQL', 'Java', '大模型(LLM)', '机器学习(ML)', 
                  '数据分析(DA)', '云原生/云计算', '大数据(BigData)', '技能总数（平均）']
print("\n表6 不同岗位类别的技能提及率对比")
print(table6)

# 热力图展示不同类别技能差异
plt.figure(figsize=(12, 6))
sns.heatmap(category_skill[display_skills], annot=True, fmt='.1f', cmap='YlGnBu', 
            cbar_kws={'label': '提及率 (%)'})
plt.title('不同岗位类别技能提及率热力图', fontsize=14)
plt.xlabel('技能')
plt.ylabel('岗位类别')
plt.tight_layout()
plt.savefig('岗位类别_技能热力图.png', dpi=300)
plt.show()
```

```python
# ============================================================
# 9. 地域分布（表7）
# ============================================================
city_dist = df['主要城市'].value_counts().head(10)
city_percent = (city_dist / len(df) * 100).round(1)
table7 = pd.DataFrame({'岗位数量': city_dist, '占比(%)': city_percent})
print("\n表7 数据类招聘岗位主要城市分布（Top 10）")
print(table7)

# 绘制城市分布条形图
plt.figure(figsize=(10, 6))
city_dist.head(10).plot(kind='bar', color='coral')
plt.title('数据类招聘岗位Top 10城市分布', fontsize=14)
plt.xlabel('城市', fontsize=12)
plt.ylabel('岗位数量', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('城市分布.png', dpi=300)
plt.show()
```

```python
# 10. 关联规则挖掘（表8, 表9）
# 准备事务数据（技能标签0/1）
skill_binary = df[skill_cols].astype(bool)

# Apriori算法：频繁项集（注意参数名为 use_colnames）
frequent_itemsets = apriori(skill_binary, min_support=0.03, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)

# 过滤长度>=2的项集（组合）
freq_2plus = frequent_itemsets[frequent_itemsets['length'] >= 2].sort_values('support', ascending=False)

# 表8：频繁技能组合（支持度≥3%）
table8 = freq_2plus[['itemsets', 'support']].head(10).copy()
table8['支持度(%)'] = (table8['support'] * 100).round(1)
table8['包含岗位数'] = (table8['support'] * len(df)).round(0).astype(int)
table8['技能组合'] = table8['itemsets'].apply(lambda x: ', '.join(list(x)))
table8_display = table8[['技能组合', '支持度(%)', '包含岗位数']]
print("\n表8 频繁技能组合（支持度≥3%）")
print(table8_display)

# 生成关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

# ===== 修复关键：强制转换数值列类型 =====
rules['support'] = pd.to_numeric(rules['support'], errors='coerce')
rules['confidence'] = pd.to_numeric(rules['confidence'], errors='coerce')
rules['lift'] = pd.to_numeric(rules['lift'], errors='coerce')
rules = rules.dropna(subset=['support', 'confidence', 'lift'])  # 丢弃转换失败的行

# 筛选提升度>1且有意义的规则
rules = rules[rules['lift'] > 1]
rules = rules.sort_values('lift', ascending=False)

# 表9：高价值关联规则（置信度≥60%）
top_rules = rules.head(8).copy()
top_rules['antecedents_str'] = top_rules['antecedents'].apply(lambda x: ', '.join(list(x)))
top_rules['consequents_str'] = top_rules['consequents'].apply(lambda x: ', '.join(list(x)))
top_rules['支持度(%)'] = (top_rules['support'] * 100).round(1)
top_rules['置信度(%)'] = (top_rules['confidence'] * 100).round(1)
top_rules['提升度'] = top_rules['lift'].round(2)

table9_display = top_rules[['antecedents_str', 'consequents_str', '支持度(%)', '置信度(%)', '提升度']]
table9_display.columns = ['前件', '后件', '支持度(%)', '置信度(%)', '提升度']
print("\n表9 高价值关联规则（置信度≥60%）")
print(table9_display)

# 可视化：支持度-置信度散点图
plt.figure(figsize=(10, 6))
scatter = plt.scatter(rules['support']*100, rules['confidence']*100,
                      c=rules['lift'], cmap='viridis', s=80, alpha=0.7)
plt.colorbar(scatter, label='提升度')
plt.xlabel('支持度 (%)', fontsize=12)
plt.ylabel('置信度 (%)', fontsize=12)
plt.title('关联规则：支持度 vs 置信度（颜色表示提升度）', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('关联规则散点图.png', dpi=300)
plt.show()
# 使用合适阈值重跑完整分析
skill_binary = df[skill_cols].astype(bool)

# 推荐阈值（可根据调试结果调整）
frequent_itemsets = apriori(skill_binary, min_support=0.005, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(len)

# 关联规则
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4)

# 转换数值类型
rules['support'] = pd.to_numeric(rules['support'], errors='coerce')
rules['confidence'] = pd.to_numeric(rules['confidence'], errors='coerce')
rules['lift'] = pd.to_numeric(rules['lift'], errors='coerce')
rules = rules.dropna()

# 筛选提升度>1
rules = rules[rules['lift'] > 1]
rules = rules.sort_values('lift', ascending=False)

print(f"最终生成 {len(rules)} 条关联规则")
print(rules[['support', 'confidence', 'lift']].head())

# 散点图
plt.figure(figsize=(10, 6))
scatter = plt.scatter(rules['support']*100, rules['confidence']*100,
                      c=rules['lift'], cmap='viridis', s=80, alpha=0.7)
plt.colorbar(scatter, label='提升度')
plt.xlabel('支持度 (%)', fontsize=12)
plt.ylabel('置信度 (%)', fontsize=12)
plt.title('关联规则：支持度 vs 置信度（颜色表示提升度）', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('关联规则散点图.png', dpi=300)
plt.show()
```

```python
# ============================================================
# 11. 额外分析：技能共现网络（可选）
# ============================================================
# 计算技能共现矩阵
co_occur = skill_binary.T.dot(skill_binary)
np.fill_diagonal(co_occur.values, 0)  # 忽略自身共现

# 找出高频共现对（支持度>5%）
threshold = 0.05 * len(df)
co_occur_pairs = co_occur.unstack().reset_index()
co_occur_pairs.columns = ['技能1', '技能2', '共现次数']
co_occur_pairs = co_occur_pairs[co_occur_pairs['技能1'] != co_occur_pairs['技能2']]
co_occur_pairs = co_occur_pairs[co_occur_pairs['共现次数'] >= threshold]
co_occur_pairs = co_occur_pairs.sort_values('共现次数', ascending=False).head(15)
print("\n高频共现技能对（共现次数≥5%总岗位数）")
print(co_occur_pairs)

```
